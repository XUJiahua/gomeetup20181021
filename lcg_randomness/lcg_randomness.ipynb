```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (20, 3)
```

```python

class LCG:
    def __init__(self, a, c, m, x0):
        self.a = a
        self.c = c
        self.m = m
        self.x0 = x0
        self.arr = []

    def next(self):
        self.x0 = (self.a * self.x0 + self.c) % self.m
        return self.x0

    def generate(self):
        self.arr = []
        for _ in range(self.m):
            self.arr.append(self.next())

    def has_dup_elem(self):
        m = {}
        for e in self.arr:
            if e in m:
                return True
            else:
                m[e] = 1

        return False


def plot_periodic(lcg):
    plt.subplot(121)
    plt.plot(list(range(lcg.m)), lcg.arr)
    plt.ylim(0, lcg.m)
    plt.grid(True)
    plt.title('periodic')


def plot_hist(lcg):
    plt.subplot(122)
    plt.hist(lcg.arr, bins=lcg.m)
    plt.title('hist')


def plot(lcg):
    lcg.generate()
    if lcg.has_dup_elem():
        print("has duplicate elements")
    plot_periodic(lcg)
    plot_hist(lcg)
```

```python
# bad randomness
plot(LCG(2, 0, 100, 3))
```

```python
# good randomness
plot(LCG(21, 3, 100, 1))
```
